## Two types of plots : Categorical Plots and Relational Plots

### Categorical Plots

The examples of categorical plots are:-

1. Count Plot 
2. Point Plot
3. Bar Plot
4. Strip Plot
5. Swarm Plot
6. Box Plot 
7. Violin Plot
8. Boxen Plot



### Function to call categorical plots : catplot()

### 1. Count Plot ( Needs only one axis specified as other is the count)

```python
sns.catplot(x="day",
            data=df,
            kind="count",
            )
```

### 2. Point Plot

```python
sns.catplot(data=df,
            x="day",
            y="tip",
            kind="point"
            )
```

### 3. Bar Plot 

The lines on the top of the bars is the confidence interval

```python
sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar"
            )
```

### 4. Strip Plot 

```python
sns.catplot(data=df,
            x="day",
            y="tip",
            kind="strip",
            )
```

### 5. Swarm Plot 

```python
sns.catplot(x="time",
            y="tip",
            data=df,
            kind="swarm",
            )
```

### 6. Box Plot 

```python
sns.catplot(x="time",
            y="tip",
            data=df,
            kind="box",
            )
```

### 7. Violin Plot 

```python
sns.catplot(x="time",
            y="tip",
            data=df,
            kind="violin",
            )
```

### 8. Boxen Plot 

```python
sns.catplot(x="time",
            y="tip",
            data=df,
            kind="boxen",
            )
```

## Customisation in seaborn

### 1. Background

```python
# SET AS WHITE GRID

sns.set_style("whitegrid")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box"
            )
```

```python
# SET AS TICKS

sns.set_style("ticks")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box"
            )
```

```python
# SET AS DARK

sns.set_style("dark")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box"
            )
```

```python
# SET AS DARK GRID

sns.set_style("darkgrid")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box"
            )
```

### 2. Category Order

```python
sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box",
            order=["Sat", "Sun", "Thur", "Fri"]
            )
```

```python
category_order = ["Sun",
                  "Sat",
                  "Fri",
                  "Thur"]

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="box",
            order=category_order
            )
```

- RdBu (Red and Blues)
- PrGn ( Purples and Green)
- pastel
- muted
- bright
- deep
- colorblind
- dark

- Greys

- 

```python
sns.set_style("white")

sns.set_palette("colorblind")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar",
            )
```

```python
# bright palette

sns.set_palette("bright")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar",
            )
```

```python
# Greys shade

sns.set_palette("Greys")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar",
            )
```

```python
# Blues shade

sns.set_palette("Blues")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar",
            )
```

### 4. Adding another dimension using hue

```python
sns.set_palette("RdBu")

sns.catplot(data=df,
            x="day",
            y="tip",
            kind="bar",
            hue="smoker"
            )
```

### 5. Making Column and Rows : subplots

```python
# Options are: "point", "bar", "strip", "swarm", "box", "violin", or "boxen

sns.set_palette("RdBu")

sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar",
            col="day",
            )

```

```python
sns.set_palette("RdBu")

sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar",
            row="day",
            )
```

```python
sns.set_palette("RdBu")

sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar",
            row="sex",
            col="day"
            )
```

### 6. Adding title

```python
sns.set_palette("RdBu")

g = sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar"
           )

g.fig.suptitle("This is my new title")
```

```python
sns.set_palette("RdBu")

g = sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar"
           )

g.fig.suptitle("This is my new title", y=1.1)
```

### 7. Adding axis labels

```python
sns.set_palette("bright")

g = sns.catplot(x="time",
            y="tip",
            data=df,
            kind="bar"
           )

g.set(xlabel="My X Label",
     ylabel = "My Y Label")
```

## Relational Plots

### Command for relational plots is relplot()


sns.scatterplot can be used but relplot is more customisable

```python
sns.relplot(x="total_bill",
            y="tip",
            data=df,
            kind="scatter")
plt.show()
```

### 1. Changing the size of the marker

```python
sns.relplot(x="total_bill",
            y="tip",
            data=df,
            kind="scatter",
            size="size"
           )
plt.show()
```

### 2. Changing the type of marker

```python
sns.relplot(x="total_bill",
            y="tip",
            data=df,
            kind="scatter",
            marker="x")
plt.show()
```

### 3. Making rows and columns / Same as the catplot()

```python
sns.relplot(x="total_bill",
            y="tip",
            data=df,
            kind="scatter",
            size="tip",
            row="smoker",
            col="day"
           )
plt.show()
```
